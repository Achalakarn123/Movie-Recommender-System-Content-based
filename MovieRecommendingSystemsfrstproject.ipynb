```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('/content/tmdb_5000_movies.csv')
credits =pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
      if counter!=3:
        L.append(i['name'])
        counter+=1
      else:
        break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
import json

def fetch_director(crew):
    try:
        # Convert string to list if needed
        if isinstance(crew, str):
            crew = json.loads(crew)  # Parse JSON string safely

        # Extract director from list
        for member in crew:
            if member.get('job') == 'Director':
                return member.get('name')

    except (json.JSONDecodeError, TypeError):  # Handle bad JSON or non-list values
        return None

    return None  # Return None if no director found
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
import ast

# Function to safely process lists of strings
def clean_list_column(column):
    return column.apply(lambda x: [i.replace(" ", "") for i in x] if isinstance(x, list) else [])

# Apply to text-based lists
movies['genres'] = clean_list_column(movies['genres'])
movies['keywords'] = clean_list_column(movies['keywords'])

# Function to clean names in dictionaries
def clean_names(column):
    return column.apply(lambda x: [i['name'].replace(" ", "") for i in x] if isinstance(x, list) and all(isinstance(i, dict) and 'name' in i for i in x) else [])

# Apply to structured data
movies['cast'] = clean_names(movies['cast'])
movies['crew'] = clean_names(movies['crew'])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
